```python
import qiskit
import numpy as np
from qiskit import QuantumCircuit, ClassicalRegister, QuantumRegister
from math import pi, sqrt
```

# CIRCUIT DEPTH
Barriers are not considered. All meas belong to the same layer of depth

```python
qc = QuantumCircuit(2)
qc.x(0)
qc.x(1)
qc.measure_all()
qc.draw('mpl')
```

```python
qc.depth()
```

```python
qc = QuantumCircuit(3)
qc.measure_all()
qc.draw('mpl')
```

```python
qc.depth()
```

```python
qc = QuantumCircuit(4)
qc.x(0)
qc.cx(0,1)
qc.s(1)
qc.cx(1,2)
qc.z(3)
qc.cx(2,3)
qc.cswap(1,2,3)
qc.draw('mpl')
```

```python
qc.depth()
```

# CIRCUIT SIZE
All gates and individual meas are considered

```python
qc = QuantumCircuit(2)
qc.x(0)
qc.x(1)
qc.measure_all()
qc.draw('mpl')
```

```python
qc.size()
```

```python
qr = QuantumRegister(3)
cr = ClassicalRegister(3)
qc = QuantumCircuit(qr, cr)
qc.x(0)
qc.h(1)
qc.x(2)
qc.cp(pi,0,1)
qc.measure_all()
qc.draw('mpl')
```

```python
qc.size()
```

# QUBIT NUMBER
Only quantum registers. Not gates, barriers, meas and classical registers

```python
qr = QuantumRegister(3)
cr = ClassicalRegister(3)
qc = QuantumCircuit(qr, cr)
qc.x(0)
qc.h(1)
qc.x(2)
qc.cp(pi,0,1)
qc.measure_all()
qc.draw('mpl')
```

```python
qc.num_qubits
```

# COUNT OPS
All the gates, barriers, single meas

```python
qr = QuantumRegister(3)
cr = ClassicalRegister(3)
qc = QuantumCircuit(qr, cr)
qc.x(0)
qc.h(1)
qc.x(2)
qc.cp(pi,0,1)
qc.measure_all()
qc.draw('mpl')
```

```python
qc.count_ops()
```

# CIRCUIT WIDTH
number of qubits, classical and meas registers.

Note: meas are created only with measure_all() method, not with measure

```python
qc = QuantumCircuit(2)
qc.x(0)
qc.h(1)

qc.measure_all()
qc.draw('mpl')
```

```python
qc.width()
```

```python
qc = QuantumCircuit(2,2)
qc.x(0)
qc.h(1)

qc.measure_all()
qc.draw('mpl')
```

```python
qc.width()
```

```python
qc = QuantumCircuit(2,2)
qc.x(0)
qc.h(1)

qc.measure([0,1],[0,1])
qc.draw('mpl')
```

```python
qc.width() #no meas registers present (not used measure_all())
```
