# Read and write in QASM file

```python
import qiskit
from qiskit import QuantumCircuit, Aer, QuantumRegister, ClassicalRegister, execute
from math import pi, sqrt
```

```python
qc = QuantumCircuit(2)
qc.x(0)
qc.y(1)
qc.z(1)
qc.draw('mpl')
```

```python
print(qc.qasm())
```

```python
qc = QuantumCircuit(2,2)
qc.cx(0,1)
qc.measure(0,1)
qc.measure(1,0)
qc.draw('mpl')
```

```python
print(qc.qasm())
```

```python
qc =QuantumCircuit(3, name='A')
qc.h(0)
qc.h(1)
qc.x(0)
qc.x(1)
qc.h(2)
qc.draw('mpl')
```

```python
print(qc.qasm())
```

```python
qc = QuantumCircuit(QuantumRegister(4,'IBM'), ClassicalRegister(4, 'EXAM'))
qc.cswap(0,1,2)
qc.swap(2,3)
qc.draw('mpl')
```

```python
print(qc.qasm())
```

```python
qc = QuantumCircuit(2)
qc.x([0,1])
qc.measure_all()
qc.draw('mpl')
```

```python
print(qc.qasm())
```
