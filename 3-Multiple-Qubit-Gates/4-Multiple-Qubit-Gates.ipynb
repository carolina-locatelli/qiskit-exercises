# MULTI QUBIT GATES

### IMPORT LIBRARIES

```python
import qiskit
import numpy as np
from qiskit import QuantumCircuit
from qiskit import QuantumRegister, ClassicalRegister, execute, Aer, BasicAer
from qiskit.visualization import visualize_transition, plot_bloch_multivector, plot_bloch_vector, plot_state_qsphere, plot_state_city, plot_histogram
from math import pi,sqrt
```

```python
Aer.backends()
```

```python
BasicAer.backends()
```

# CX GATE

```python
QC = QuantumCircuit(2)
QC.x(0)
QC.x(1)
QC.cx(0,1)
QC.draw('mpl')
```

### Bloch sphere
NB: the Bloch sphere will not represent the state in little endian notation. only the statevector will do it

```python
#Initialize the backend (without measurements)
backend = Aer.get_backend('statevector_simulator')

#Execute circuit
result = execute(QC, backend).result().get_statevector(QC, decimals =3)

#Print result
print('\n Quantum state is:', result)

#Plot Bloch sphere
plot_bloch_multivector(result)
```

### State city

```python
plot_state_city(QC, color=['purple','crimson'], title='3D representation of CX gate')
```

### Q sphere

```python
plot_state_qsphere(QC)
```

### Unitary operator

```python
#Initialize the backend
backend = Aer.get_backend('unitary_simulator')

#Execute circuit
x_unitary=execute(QC,backend)
x_unitary.result().get_unitary(QC, decimals = 3)
```

### Histogram

```python
QC = QuantumCircuit(2,2)
QC.x(0)
QC.x(1)
QC.cx(0,1)
QC.measure(0,0)
QC.measure(1,1)
QC.draw('mpl')
```

```python
#Set the backend
backend = Aer.get_backend('qasm_simulator')

#Execute circuit
job = execute(QC,backend,shots=10000)

#Get result
result = job.result()

#Count
counts=result.get_counts(QC)

#Print state
print('\n Quantum state is:', counts)
```

```python
plot_histogram(counts)
```

# READ AND WRITE IN QASM FILES

The circuits will be converted to Assembly language

```python
qc =QuantumCircuit(2)
qc.y(0)
qc.draw('mpl')
```

```python
print(qc.qasm())
```

```python
qr =QuantumRegister(2)
cr=ClassicalRegister(2)
qc=QuantumCircuit(qr,cr)
qc.z(0)
qc.cnot(0,1)
qc.measure_all()
qc.draw('mpl')
```

```python
print(qc.qasm()) # Barriers and measurements are included too
```
