# Bloch sphere
ORIGINAL SYNTAX

plot_bloch_vector(bloch, title='', ax=None, figsize=None, coord_type='cartesian' or 'radians')

--> Bloch can be cartesian and polar

```python
import qiskit
from qiskit.visualization import plot_bloch_vector
from math import pi, sqrt
```

```python
plot_bloch_vector([1,0,0], coord_type='cartesian')
```

```python
plot_bloch_vector([0,1,0], title='New sphere')
```

```python
plot_bloch_vector([0,0,1])
```

```python
plot_bloch_vector([1,1,0])
```

```python
plot_bloch_vector([1,0,1])
```

```python
plot_bloch_vector([0,0,-1])
```

```python
plot_bloch_vector([-1,0,-1], coord_type='cartesian')
```

```python
plot_bloch_vector([-1,-1,0], coord_type='cartesian')
```

```python
plot_bloch_vector([pi,pi,pi], coord_type='radians')
```
