# Exercises

### IMPORTING LIBRARIES

```python
import qiskit
import numpy as np
from qiskit import Aer, QuantumCircuit, execute, QuantumRegister, ClassicalRegister, BasicAer
from qiskit.visualization import plot_bloch_multivector #blochsphere
from qiskit.visualization import plot_histogram #2D histogram
from qiskit.visualization import plot_state_city #3D histogram
from qiskit.visualization import plot_state_qsphere, plot_bloch_vector
from qiskit.visualization import visualize_transition
from math import pi, sqrt
```

### ACCESSING AER AND BASIC BACKENDS

```python
Aer.backends() #C++ code in backend
```

```python
BasicAer.backends() #Python code in backend
```

*StatevectorSimulator: plot_bloch_multivector, plot_state_city, plot_state_qsphere = 3D visualization*

*QasmSimulator: plot_histogram = 2D visualization*

*UnitarySimulator = matrix representation*

### PAULI X GATE

```python
qc = QuantumCircuit(1)
qc.x(0)
qc.draw('mpl')
```

### Bloch sphere

```python
#Initialize the backend (without measurements)
backend = Aer.get_backend('statevector_simulator')

#Execute the circuit
result = execute(qc, backend).result().get_statevector(qc, decimals = 3)

#Print the result
print('\n Quantum state is:', result)

#Plot Bloch sphere
plot_bloch_multivector(result)
```

### Plot state city

```python
plot_state_city(qc, color=['purple', 'crimson'], title='3D representation of Pauli X gate')
```

### Qsphere

```python
plot_state_qsphere(qc)
```

### Unitary operator

```python
#Initialize the backend as unitary simulator
backend = Aer.get_backend('unitary_simulator')

#Execute the circuit
x_unitary = execute(qc, backend)
x_unitary.result().get_unitary(qc, decimals = 3)
```

### Plotting histogram
In order to plot the histogram, we need to measure the circuit!

```python
qc = QuantumCircuit(2,2)
qc.x(0)
qc.x(1)
qc.measure(0,0)
qc.measure(1,1)
qc.draw('mpl')
```

```python
#Set the backend simulator
backend = Aer.get_backend('qasm_simulator')

#Execute the circuit
job = execute(qc, backend, shots = 10000)

#Get result from job
result = job.result()

#Return count
counts = result.get_counts(qc)

#Print the state
print('\n The quantum state is:', counts)
```

```python
#Plot histogram
plot_histogram(counts)
```
